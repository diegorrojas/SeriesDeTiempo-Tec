**Introducción**
 
Una serie de tiempo es basicamente una sucesión de datos en forma cronologica que pueden ser utilizados para optimizar o predecir otros datos. En estas series de tiempo puede variar su patrón.  Para realizar una predicción se utiliza un modelo Arima en el cual primero se hace una diferenciación de los valores, despues se hace un analisis ARMA hasta un rango de 5 para poder generar una predicción de modelo ARIMA con los valores mas pequeños de AIC y BIC, despues generando los pronosticos. Estos pronosticos se basan en información pasada. 

**Series Elegidas**
 
**Inflación en México**

Se usaron datos de la inflación en México mensual desde el año 2000 hasta 2023  y con ello se hizo un pronostico para meses subsecuentes.
 
La inflación viene siendo el incremento de los precios en una economia durante un periodo de tiempo, esto reduce el poder adquisitvo de la gente. En México la inflación lo mide el Indice Nacional de Precios al Consumidor (INPC) y publica los datos la INEGI. El INPC es practicamente un indice el cual se dedica a medir las variaciones de los precios en 1 mes de 8 categorias diferentes de productos para poder medir la inflación del país. 
 
En cuanto a las empresas la inflación puede afectar fuertemente a las empresas y su planeación financiera. En primer lugar, la mayor amenaza de la inflación es el incremento de costos, la inflación hace que las cosas cuesten mas y por lo tanto suben los costos en una empresa generando menos ganacias. Otra amenaza es que las ventas de una empresa pueden bajar, esto debido a que la inflación puede generar una baja en el poder adquisitvo de la gente.

**Acciones WALMEX**

Con las acciones de Walmex desde el 2017 de valor de cierre ajustado diario, se obtuvieron mas de 1500 datos y se hizo un pronostico.

Una acción es practicamente un pedazo de la empresa, comprar una acción es basicamente comprar una parte de la empresa y en la cual el precio varia segun la demanda cierta empresa tenga. Una empresa recauda dinero con acciones el momento que las vende por primera vez en el mercado, tambien puede generar despues mas acciones para venderlas. Algunas de estas empresas con acciones ofrecen dividendos a los accionistas, un dividendo es practicamente una repartición de ganancias entre los dueños, en este caso los accionistas reciben un pago cada varios meses.

Una acción ordinaria es una acción ls cual otorga propiedad parcial de una empresa, estos propietarios de acciones ordinarias pueden participar en la empresa y sus actividades diarias como en la toma de decisiones  y recibir dividendos en caso de que los ofrezcan, aunque es mas riesgoso contar con esta clase de acción ya que en que caso de que quiebre la empresa estos accionistas tienen menor prioridad en la repartición de activos.

A una empresa le es importante saber el precio de su acción ya que con este puede saber muchas cosas importantes. Con el precio de su acción puede averiguar cuanto puede llegar a valer su empresa. También puede ver el rendimiento que ha tenido su empresa y compararse con la competencia para ver como van.
